```python
stock_data = pd.read_excel(r'data\stock_data2.xlsx')
stock_data['sold_ratio'] = stock_data['total_sold'] / stock_data['initial_stock']
stock_data
```

```python
dead_stock_items = stock_data[stock_data['sold_ratio'] < 0.6]['product_name'].tolist()
dead_stock_items
```

```python
 basket_data = data.groupby('transaction_id')['product_name'].apply(list).tolist()
 basket_data
```

```python
 Te = TransactionEncoder()
 te_ary = Te.fit(basket_data).transform(basket_data)
 basket = pd.DataFrame(te_ary, columns=Te.columns_)
 basket
```

```python
# basket = data.groupby(['transaction_id', 'product_name'])['quantity'].sum().unstack().fillna(0)
# basket = basket.applymap(lambda x: 1 if x > 0 else 0)
# basket
```

```python
 frequent_itemsets = apriori(basket, min_support=0.1, use_colnames=True)
 frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules = rules[rules['confidence'] >= 0.05] 
rules = rules[rules['support'] >= 0.05]
bundling = rules.sort_values(by='confidence', ascending=False)
bundling.head(10)
```

```python
bundling['consequents'] = bundling['consequents'].apply(lambda x: [p for p in x if p in dead_stock_items])
bundling = bundling[bundling['consequents'].map(len) > 0]
bundling = bundling[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
bundling = bundling.sort_values(by='confidence', ascending=False)
# g = bundling[bundling['antecedents'] == frozenset(['Running Shorts'])]
bundling
```

```python
best_rules_idx = bundling.groupby('antecedents')['confidence'].idxmax()
best_recommendations = bundling.loc[best_rules_idx]

best_recommendations = best_recommendations[['antecedents', 'consequents', 'support', 'confidence', 'lift']].reset_index(drop=True)

best_recommendations
```

```python
input_products = ['Basketball','Running Shorts']

matched_best_rules = best_recommendations[
    best_recommendations['antecedents'] == frozenset(input_products)
]

# Extract the recommended deadstock items from the consequents of these matched best rules
recommended_dead_stock = set()
for _, row in matched_best_rules.iterrows():
    recommended_dead_stock.update(row['consequents'])

print("ðŸŽ¯ Recommended Dead Stock Bundles:", list(recommended_dead_stock))
```
