```python
outlook=['sunny','sunny','overcast','rainy','rainy','rainy',
         'overcast','sunny','sunny','rainy','sunny',
         'overcast','overcast','rainy']
temperature=['hot','hot','hot','mild','cool','cool','cool','mild',
             'cool','mild','mild','mild','hot','mild']
humidity=['high','high','high','high','normal','normal',
          'normal','high','normal','normal','normal','high','normal','high']

wind=['weak','strong','weak','weak','weak','strong','strong','weak',
      'weak','weak','strong','strong','weak','strong']

play=['no','no','yes','yes','yes','no','yes','no','yes','yes',
      'yes','yes','yes','no']
```

```python
#convert string data into numerical data ====>encoding techniques

#1--label encooding 
#2--one hot encoding 
#3--get dummies
```

```python
#1--label Encoding===> Machine automaticaly assign a values in every strings
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
#convert string into numerical data
outlook_encoder=le.fit_transform(outlook)
temperature_encoder=le.fit_transform(temperature)
humidity_encoder=le.fit_transform(humidity)
wind_encoder=le.fit_transform(wind)
play_encoder=le.fit_transform(play)
outlook_encoder
#2==sunny
#1==raniy
#0==overcast
```

```python
temperature_encoder
#2==mild
#1==hot
#0==cool
```

```python
humidity_encoder
#1==normal
#0==high
```

```python
wind_encoder
#1==weak
#0==strong
```

```python
play_encoder
```

```python
#zip
feature=list(zip(outlook_encoder,temperature_encoder,humidity_encoder,wind_encoder))
feature
```

```python
#no need to use normalization and spliting
#model creation step
# there are 3 types of classifier

#1--multinomial   ===>discreate dataset(like string data sets)
#2--Gaussian      ===>continues data(numarical data)
#3--bernolli      ===>combained data(both string and numerical data)

from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(feature,play)
print(model.predict([[2,0,0,0]]))
```
