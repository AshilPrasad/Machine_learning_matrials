```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/customer5.txt",header=None)
df
```

```python
df.columns=["id","fname","lname","age","prof","country","salary"]
df
```

```python
#question-1
df1=df.groupby("prof")["prof"].count()
df1
```

```python
import matplotlib.pyplot as plt
plt.pie(df1)
plt.legend(df1)
```

```python
#question-2
df2=df.groupby('country')['country'].count()
df2
```

```python
plt.plot(df2,'*--r',ms=20,mfc='b')
plt.xlabel("country")
plt.ylabel("count")
plt.title("country and its count")
```

```python
#questoin-3
df3=df.groupby("prof")["salary"].max()
df3
```

```python
plt.plot(df3,'*-g')
plt.xlabel("prof")
plt.ylabel("salary")
plt.title("prof and salary")
```

```python
#question-4
df4=df.groupby("prof")["salary"].sum()
df4
```

```python
plt.plot(df4,linewidth=3)
plt.xlabel("prof")
plt.ylabel("salary")
plt.title("prof and its salary")
```
