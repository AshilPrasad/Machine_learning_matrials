```python
x=[150,145,178,190]
y=[45,56,78,69]
plt.plot(x,y)
plt.xlabel("height in cm")
plt.ylabel("weight in cm")
plt.title("height and weight graph")
```

```python
x=[150,145,178,190]
y=[45,56,78,69]
font1={"family":'serif',"color":"red","size":20}
font2={"family":'serif',"color":"blue","size":30}
plt.plot(x,y)
plt.xlabel("height in cm",fontdict=font1)
plt.ylabel("weight in cm",fontdict=font1)
plt.title("height and weight graph",)
plt.title("height and weight graph",fontdict=font2)
plt.grid(axis="x")
```

```python
#velocity-displacement grapg

x=[12,34,23,56,34,67]
y=[23,34,67,78,89,100]
plt.plot(x,y,'p-g',mfc="red",linewidth=5)
font1={"family":"serif","color":"green","size":10}
font2={"family":"serif","color":"yellow","size":15}
plt.xlabel("velocity",fontdict=font1)
plt.ylabel("displacement",fontdict=font1)
plt.title("velocity - displacement grpah",fontdict=font2,loc="right")
plt.grid(color="red",linestyle="dashed",linewidth="2") #grid function
```
